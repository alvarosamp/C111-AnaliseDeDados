```python
import pandas as pd
```

```python
seriesAno1 = pd.Series({'Java': 16.25, 'C': 16.04, 'Python': 9.85})
seriesAno2 = pd.Series({'C': 16.21, 'Python': 12.12, 'Java': 11.68})
```

```python
totalAno1 = seriesAno1.sum()
totalAno2 = seriesAno2.sum()

print("Total Ano 1:", totalAno1)
print("Total Ano 2:", totalAno2)
```

```python
variacao = seriesAno2 - seriesAno1
print(variacao)
```

```python
crescimento = variacao[variacao > 0]
print(crescimento)
```

```python
ano4 = seriesAno2 + 2 * variacao
print(ano4)
```

```python
mais_popular = ano4.nlargest(1)
print(mais_popular)
```
