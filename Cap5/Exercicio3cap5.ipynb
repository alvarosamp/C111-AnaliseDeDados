```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/paises (1).csv', sep =';')
df.head()
```

```python
#Paises da oceania
oceania_countries = df[df['Region'].str.contains('OCEANIA', case=False, na=False)]
print("Países da Oceania:")
print(oceania_countries[['Country']])

# Quantos países são da Oceania
print(f"\nTotal de países da Oceania: {len(oceania_countries)}")
```

```python
#Questao dois
mais_pop = df.loc[df['Population'].idxmax()]
print("\nPaís com a maior população:")
print(f"Nome: {mais_pop['Country']}, Região: {mais_pop['Region']}")
```

```python
# Agrupar por regiões e calcular a média de alfabetização
literacy = df.groupby('Region')['Literacy (%)'].mean().reset_index()
print("\nMédia de alfabetização por região:")
print(literacy)
```

```python
# Filtrar países sem costa marítima
sem_costa = df[df['Coastline (coast/area ratio)'] == 0]
sem_costa.to_csv('noCoast.csv', index=False)  # Salvar o arquivo
print("\nPaíses sem costa marítima salvos em 'noCoast.csv'")
```

```python
# Função para classificar os países com base na taxa de mortalidade
df['Humanitarian Help'] = df['Deathrate'].apply(lambda x: 'Balanced' if x < 9 else 'Urgent')

# Verificar a atualização no dataset
print("\nDataset atualizado com a coluna 'Humanitarian Help':")
print(df[['Country', 'Deathrate', 'Humanitarian Help']].head())
```
