```python
import pandas as pd

df = pd.DataFrame(
    {
        'W': [10, 29, 30, 43, 37],
        'X': [37, 26, 9, 41, 48],
        'Y': [16, 30, 10, 37, 12],
        'Z': [1, 49, 1, 17, 25]
    },
    index=['A', 'B', 'C', 'D', 'E']
)

print(df)
```

```python
media_x_menor_30 = df[df['X'] < 30]['X'].mean()
print( media_x_menor_30)
```

```python
media_linha_D = df.loc['D'].mean()

# Soma da linha E (índice 4)
soma_linha_E = df.iloc[4].sum()

print("Média da linha D:", media_linha_D)
print("Soma da linha E:", soma_linha_E)
print()
```

```python
# Selecionando linhas A, C, E e colunas X, Y
slice_df = df.loc[['A', 'C', 'E'], ['X', 'Y']]

print("8) Slice (linhas A, C, E | colunas X, Y):")
print(slice_df)
print()

# Soma por linha
soma_linhas = slice_df.sum(axis=1)
print("Soma por linha:")
print(soma_linhas)
print()

# Soma por coluna
soma_colunas = slice_df.sum(axis=0)
print("Soma por coluna:")
print(soma_colunas)
```
