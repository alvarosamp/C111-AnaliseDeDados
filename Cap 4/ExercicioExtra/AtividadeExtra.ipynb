```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/paises.csv', sep =';')
df.head()
```

```python
#Faça um slicing no dataset para mostrar apenas o País (Country), Região
#(Region), População (Population) e Area (Area (sq. mi.)) dos países contidos
#nele
colunas = df[['Country', 'Region', 'Population', 'Area (sq. mi.)']]
colunas.head()
```

```python
#Diferentes regioes do planeta
df['Region'].value_counts()
```

```python
#Mostre qual a taxa média de alfabetização (Literacy (%)) do planeta segundo
media = df['Literacy (%)'].mean()
print(media)
```

```python
#Conte quantos países são da América do Norte (NORTHERN AMERICA)
northern_america_countries = df[df['Region'].str.strip() == 'NORTHERN AMERICA']
num_northern_america = len(northern_america_countries)
print(f"Número de países na NORTHERN AMERICA: {num_northern_america}")
```

```python
#Encontre qual país da América do Sul e Caribe (LATIN AMER. & CARIB)
df['Region'] = df['Region'].str.strip()
america_latina_caribe = df[df['Region'] == 'LATIN AMER. & CARIB']
pais_mais_rico_america_latina = america_latina_caribe.loc[america_latina_caribe['GDP ($ per capita)'].idxmax()]
print(f"{pais_mais_rico_america_latina['Country']} com GDP de {pais_mais_rico_america_latina['GDP ($ per capita)']}$")
```
