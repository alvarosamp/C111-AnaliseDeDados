```python
import numpy as np
#Variando de 1 a 7 
array1 = np.random.randint(1, 7, size=(np.random.randint(1, 7), np.random.randint(1, 7)))
print(array1.shape)
print('-----')
mult = array1.shape[0] * array1.shape[1]
print(mult)
print('-----')
if mult % 2 == 0:
    print('A matriz pode ser transformada em um vetor par')
else:
    print('A matriz pode ser transformada em um vetor ímpar')
```

```python
np.random.seed(10)

matriz = np.random.randint(1, 51, size=(4, 4))
print("Matriz Gerada:")
print(matriz)

media_linhas = np.mean(matriz, axis=1)  
media_colunas = np.mean(matriz, axis=0)  
print("\nMédia de cada linha:")
print(media_linhas)
print("\nMédia de cada coluna:")
print(media_colunas)


maior_media_linhas = np.max(media_linhas)
maior_media_colunas = np.max(media_colunas)
print("\nMaior valor das médias das linhas:", maior_media_linhas)
print("Maior valor das médias das colunas:", maior_media_colunas)


unique, counts = np.unique(matriz, return_counts=True)
aparicoes = dict(zip(unique, counts))

numeros_2_vezes = {k: v for k, v in aparicoes.items() if v == 2}
print("\nNúmeros que aparecem 2 vezes:")
print(numeros_2_vezes)
```
