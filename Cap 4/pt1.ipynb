```python
import numpy as np

array1 = np.ones(8)
array2 = np.random.randint(0, 9, 8)

#Salvando no terceiro
array3 = np.add(array1, array2)

if np.sum(array3) >=40:
    array3 = np.reshape(array3,(4,2))
else:
    array3 = np.reshape(array3,(2,4))
print(array3)

```

Exercicio 3

```python
array0 = np.zeros([2,2])
array0[np.random.randint(0,2), np.random.randint(0,2)] = 1 #Pegando as posições aleatórias de linha e coluna
print(array0)
tentativas = 0
while(tentativas  <3):
    print('Selecione uma linha e uma colunas')
    linha = int(input('Linha: '))
    coluna = int(input('Coluna: '))
    print('-----')
    print(linha, coluna)
    if array0[linha, coluna] == 1:
        tentativas +=1
        print('Congratulations! You beat the game! :)')
        break
    else:
        tentativas +=1
        if tentativas == 3:
            print('Game Over! :( Try again!') 
        
            
```
