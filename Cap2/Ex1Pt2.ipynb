```python
#ex1
distancia = float(input("Digite a distância percorrida em km: "))

preco_km = 0.5 if distancia <= 200 else 0.45
valor = distancia * preco_km

print(f"O valor da corrida fica: R$ {valor:.2f}")
```

```python
#ex2
num = int(input("Digite um numero entre 1000 e 9999: "))

if 1000 <= num <= 9999:
    unidade = num % 10
    dezena = (num // 10) % 10
    centena = (num // 100) % 10
    milhar = num // 1000

    print(f"Unidade: {unidade}")
    print(f"Dezena: {dezena}")
    print(f"Centena: {centena}")
    print(f"Milhar: {milhar}")
else:
    print("Digite um numero valido")
```

```python
import math
num = float(input("Digite um numero decimal: "))
print(f'Raiz de num : {math.sqrt(num)}')
#função teto(aprox p cima)
print(f' Função teto : {math.ceil(num)}')
#Função chao
print(f' Funcao chao : {math.floor(num)}')
#Parte inteira
print(f'Parte inteira : {int(num)}')
```
